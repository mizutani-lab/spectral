```python
import networkx as nx
import numpy as np
from numpy.linalg import eig
import sympy as sp

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)
```

```python
import easy_plotly as ep
```

```python
def plot_eig_vec(G):
    L = nx.laplacian_matrix(G).toarray()
    w, v = eig(L)
    idx = np.argsort(w)
    w, v = w[idx], v[:, idx]

    fig = ep.PlotlyFigure('./scatter.yml', x_title='$\psi_2$', y_title='$\psi_3$')
    fig.add_scatter(v[1], v[2], marker_color=list(range(len(G))), marker_colorscale="RdBu")
    fig.to_png(show=True)
    
    # fig2 = ep.PlotlyFigure('./line.yml', x_title='vertex', y_title='eigenvalues')    
    fig2 = ep.PlotlyFigure('./scatter.yml', x_title='vertex', y_title='eigenvalues')
    fig2.add_scatter(list(range(len(G))), w, marker_color=list(range(len(G))), marker_colorscale="RdBu")
    fig2.to_png(show=True)
```

```python
plot_eig_vec(nx.cycle_graph(200))
```

```python
plot_eig_vec(nx.path_graph(200))
```
